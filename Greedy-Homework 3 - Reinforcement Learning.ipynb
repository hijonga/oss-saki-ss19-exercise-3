```python
start=datetime.datetime.now()

curr_state = ['0','0','0','0','0','0']
last_state = ['0','0','0','0','0','0']
price = [-1,-2,-2,-3,-3,-4]
not_possible_in = -100
not_possible_out = -100
no_storage = 0
no_restorage = 0
total_cost = 0
size_storage = 6
storage_places = range(1,(size_storage+1))

# load order
command = pd.read_csv("warehousetraining_2x3.txt", delimiter='\t', names=["instr", "next_color"])
print(len(command))

for move in command.itertuples():
  print("command:" + move.instr +' '+ move.next_color)
  action_impossible = True
  for place in storage_places:
    # find nearest position with right characteristics
    # update current state
    if move.instr == 'store':
      if(curr_state [place-1] == '0'):
        curr_state [place-1] = move.next_color
        total_cost = total_cost + price[place-1]
        print('reward: '+str(total_cost))
        action_impossible = False
        break
    elif move.instr == 'restore':
      if(curr_state [place-1] == move.next_color):
        curr_state [place-1] = '0'
        total_cost = total_cost + price[place-1]
        print('cost: '+str(total_cost))
        action_impossible = False
        break


  if(action_impossible):
      if move.instr == 'store':
        print ('storage refused')
        no_storage =no_storage +1
        total_cost = total_cost + not_possible_in 
      elif move.instr == 'restore':
        print ('restorage not possible')
        no_restorage =no_restorage +1
        total_cost = total_cost + not_possible_out 
  print(curr_state)

print('TOTAL REWARD:    ' + str(total_cost))
print(datetime.datetime.now()-start)
```
